```python
# Extract unique artists and genres
all_artists = list(artists_data.keys())
all_genres = sorted(list(set([genre for genres in artists_data.values() for genre in genres])))

print(f"Total unique artists: {len(all_artists)}")
print(f"Total unique genres: {len(all_genres)}")
print(f"\nGenres: {all_genres}")
```

```python
# Create the dataset with one-hot encoding
data = []

for song_name, artist in songs_library:
    full_song_name = f"{song_name} - {artist}"
    
    # Initialize row with song name
    row = {'Song_Name': full_song_name}
    
    # One-hot encode artists
    for a in all_artists:
        row[f'Artist_{a}'] = 1 if a == artist else 0
    
    # One-hot encode genres based on artist's genres
    artist_genres = artists_data[artist]
    for g in all_genres:
        row[f'Genre_{g}'] = 1 if g in artist_genres else 0
    
    data.append(row)

# Create DataFrame
df = pd.DataFrame(data)

print(f"Dataset shape: {df.shape}")
print(f"\nFirst few rows:")
df.head()
```

```python
print("Column names:")
print(df.columns.tolist())
print(f"\nTotal columns: {len(df.columns)}")
print(f"Song_Name column: 1")
print(f"Artist columns: {len(all_artists)}")
print(f"Genre columns: {len(all_genres)}")
```

```python
# Save to CSV
output_filename = 'Data.csv'
df.to_csv(output_filename, index=False)
print(f"Dataset saved as: {output_filename}")
print(f"Total songs: {len(df)}")
```
